Feature Scaling


```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
df['species']=label_encoder.fit_transform(df['species'])
df['sex']=label_encoder.fit_transform(df['sex'])
df['island']=label_encoder.fit_transform(df['island'])
df.head()
```

```python
dummies = pd.get_dummies(df[['island','species']],drop_first=True)
df_to_be_scaled = df.drop(['island','species'],axis=1)
target = df_to_be_scaled.sex
df_feat= df_to_be_scaled.drop('sex',axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df_feat)
df_scaled = scaler.transform(df_feat)
df_scaled = pd.DataFrame(df_scaled,columns=df_feat.columns[:4])
df_preprocessed = pd.concat([df_scaled,dummies,target],axis=1)
df_preprocessed.head()
```

```python
df_preprocessed=df_preprocessed.dropna(how='any',axis=0)
df_preprocessed.head()
```

```python
from sklearn.model_selection import train_test_split

train,test=train_test_split(df_preprocessed,test_size=0.2,random_state=10)
```
